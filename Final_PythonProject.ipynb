# Data Cleaning - Handling Duplicates and missing values


```python
df_sales.isnull().sum()
```

```python
df_products.isnull().sum()
```

```python
# Handle null values in 'city' and 'delivery_status' by dropping rows
# Q.For the city and delivery_status columns, remove the records (rows) tha contain NaN values.
df_sales.dropna(subset=['city','delivery_status'],inplace = True)
```

```python
df_sales[["city","delivery_status"]].isnull().sum()
```

```python
df_sales.isnull().sum()
```

```python
# Handle null values in 'delivery_time_mins' by filling with the mean
# Q.For the delivery_time_mins column, fill missing values with the mean of the column.

df_sales["delivery_time_mins"]=df_sales["delivery_time_mins"].fillna(df_sales["delivery_time_mins"].mean()) 

```

```python
df_sales["delvery_time_mins"].isnull().sum()
```

```python
df_sales.isnull().sum()
```

```python
# Check for duplicate records

df_sales.duplicated().sum()
```

```python
# Remove duplicate records 

df_sales.drop_duplicates(inplace=True)
```

```python
# Convert 'order_date' to datetime objects

df_sales["order_date"] = pd.to_datetime(df_sales["order_date"])
```

```python
df_sales["order_date"]
```

# Data Analysis - Aggregations and Grouping

```python
# min amount
min_amount = df_sales["total_amount"].min()
print(min_amount)
```

```python
# max amount
max_amount = df_sales["total_amount"].max()
print(max_amount)
```

```python
# avg amount
avg_amount = df_sales["total_amount"].mean()
print(avg_amount)
```

```python
# Top 5 products by total sales amount

top_5products = df_sales.groupby("product_id")["total_amount"].sum().sort_values(ascending = False).head(5)
print(top_5products)
```

```python
df_sales.columns
```

```python
# Merge With Product Details to get Product Names.

top_products_name = top_5products.reset_index().merge(df_products,on='product_id')
print(top_products_name)
```

```python
# Total sales By City

total_sales = df_sales.groupby("city")["total_amount"].sum()
print(total_sales)
```

```python
# Average Delivery Time by city
average_delivery_time = df_sales.groupby('city')['delivery_time_mins'].mean()
print(average_delivery_time)
```

```python
# Sales Trend over Time (monthly sales)

#creating month column
df_sales["month"] = df_sales['order_date'].dt.to_period('M')

#Group by
monthly_sales_trends = df_sales.groupby('month')['total_amount'].sum()
print(monthly_sales_trends)
```

```python
# sales By Product category

df_products.columns
```

```python
sales_by_category = df_sales.merge(df_products, on="product_id")
sales_by_category = sales_by_category.groupby("category")["total_amount"].sum().sort_values(ascending=False)
print(sales_by_category)
```
